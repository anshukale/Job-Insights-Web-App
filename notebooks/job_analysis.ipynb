```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("D:\IIT Gandhinagar\Off-Campus\Projects\Job Insights Web App\data\job_postings.csv")

# Display basic info
print(df.info())

# Show first 5 rows
df.head()
```

```python
# Count occurrences of each job title
top_jobs = df['Job Title'].value_counts().head(10)

# Plot top job roles
plt.figure(figsize=(10, 5))
sns.barplot(x=top_jobs.values, y=top_jobs.index, palette="viridis")
plt.xlabel("Number of Postings")
plt.ylabel("Job Title")
plt.title("Top 10 Most In-Demand Job Titles")
plt.show()
```

```python
# Count occurrences of work types (Remote, Hybrid, On-Site)
work_type_counts = df['Work Type'].value_counts()

# Plot work type distribution
plt.figure(figsize=(6, 4))
sns.barplot(x=work_type_counts.index, y=work_type_counts.values, palette="coolwarm")
plt.xlabel("Work Type")
plt.ylabel("Number of Postings")
plt.title("Work Type Distribution (Remote vs On-Site)")
plt.show()
```

```python
# Count job postings per country
top_countries = df['Country'].value_counts().head(10)

# Plot top hiring locations
plt.figure(figsize=(10, 5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="Blues_r")
plt.xlabel("Number of Postings")
plt.ylabel("Country")
plt.title("Top 10 Hiring Countries")
plt.show()
```
